Here we are importing and plotting population data 

```python
#Import "Population_Data.csv" into a pandas Dataframe
import pandas as pd 
```

```python
pop_data = pd.read_csv("Population_Data.csv") 
pop_data.head() 
```

```python
import matplotlib.pyplot as plt
# Remove all other lines at 276, only keep the green dashed line
plt.axhline(y=276, color='green', linestyle='--', linewidth=2)
plt.yticks(list(plt.yticks()[0]) + [276], labels=[*map(str, plt.yticks()[0]), '276'])
plt.plot(pop_data["Days"], pop_data["Population"], color='black', marker='o', linestyle='-')
plt.grid(True)
plt.plot(pop_data["Days"], pop_data["Population"], marker='o', color='black', linestyle='-')
plt.xlim(left=0)
plt.ylim(bottom=0)
plt.plot(pop_data["Days"], pop_data["Population"])  
plt.xlabel("Days")
plt.ylabel("Population Size")       
plt.title("Population Growth Over Time")
plt.show()
```

```python
#Average the population from days 17 to days 60 
avg_pop = pop_data["Population"][17:61].mean() 
print(f"The average population from days 17 to 60 is {avg_pop:.2f}")
```
