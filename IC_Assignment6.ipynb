```python
import numpy as np
import matplotlib.pyplot as plt

def beetle_population(years, initial_pop1, initial_pop2):
    # Transition matrix for age-structured beetle population
    A = np.array([[0, 1, 5],      # Offspring produced by 2nd and 3rd year beetles
                  [0.3, 0, 0],    # Survival from 1st to 2nd year
                  [0, 0.5, 0]])   # Survival from 2nd to 3rd year
    # Initial population vector: [first year, second year, third year]
    n_t = np.array([initial_pop1, initial_pop2, 0])
    # Lists to store populations for each year
    first_year_pop = []
    second_year_pop = []
    third_year_pop = [] 
    total_pop = []
    # Calculate populations for each year
    for year in range(years):
        first_year_pop.append(n_t[0])
        second_year_pop.append(n_t[1])
        third_year_pop.append(n_t[2])
        total_pop.append(sum(n_t))
        n_t = np.dot(A, n_t)
    return first_year_pop, second_year_pop, third_year_pop, total_pop

# Run the model for 50 years with 500 beetles in first and second year
first_year_pop, second_year_pop, third_year_pop, total_pop = beetle_population(50, 500, 500)

# Plot the results
plt.plot(first_year_pop, label='First Year Beetles')
plt.plot(second_year_pop, label='Second Year Beetles')
plt.plot(third_year_pop, label='Third Year Beetles')
plt.plot(total_pop, label='Total Population', linestyle='--')
plt.xlabel('Years')
plt.ylabel('Population')
plt.title('Beetle Population Over Time')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def beetle_population(years, initial_pop1, initial_pop2, matrix):
    n_t = np.array([initial_pop1, initial_pop2, 0])
    first_year_pop = []
    second_year_pop = []
    third_year_pop = []
    total_pop = []
    for year in range(years + 1):  # include year 0
        first_year_pop.append(n_t[0])
        second_year_pop.append(n_t[1])
        third_year_pop.append(n_t[2])
        total_pop.append(sum(n_t))
        n_t = np.dot(matrix, n_t)
    return first_year_pop, second_year_pop, third_year_pop, total_pop


years = 50
initial_pop1 = 500
initial_pop2 = 500

# No-trap scenario matrix
A_no_trap = np.array([
    [0, 1, 5],    # Offspring produced by 2nd and 3rd year beetles
    [0.3, 0, 0],  # Survival from 1st to 2nd year
    [0, 0.5, 0]   # Survival from 2nd to 3rd year
])

# Trap scenario matrix (5% reduction in survival for 2nd and 3rd year)
A_trap = np.array([
    [0, 1, 5],
    [0.3, 0, 0],
    [0, 0.475, 0]  # 0.5 * 0.95 = 0.475
])

# Run both scenarios
fy_no_trap, sy_no_trap, ty_no_trap, tot_no_trap = beetle_population(years, initial_pop1, initial_pop2, A_no_trap)
fy_trap, sy_trap, ty_trap, tot_trap = beetle_population(years, initial_pop1, initial_pop2, A_trap)

# Plot results
plt.figure(figsize=(10,6))
plt.plot(tot_no_trap, label='Total (No Trap)', color='blue')
plt.plot(tot_trap, label='Total (With Trap)', color='red', linestyle='--')
plt.plot(fy_no_trap, label='First Year (No Trap)', color='lightblue', alpha=0.5)
plt.plot(sy_no_trap, label='Second Year (No Trap)', color='green', alpha=0.5)
plt.plot(ty_no_trap, label='Third Year (No Trap)', color='orange', alpha=0.5)
plt.xlabel('Years')
plt.ylabel('Population')
plt.title('Beetle Population Projection (20 Years)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Print final populations for both scenarios
print(f"Final population after 50 years (No Trap): {int(tot_no_trap[-1])}")
print(f"Final population after 50 years (With Trap): {int(tot_trap[-1])}")
```

1. What differnece do you see between the teo plots?  
The No Trap scenario (blue line) shows a much higher and steadily increasing beetle population over time, while the With Trap scenario (red dashed line) grows more slowly and levels off at a lower population. The traps cause a noticeable reduction in the total beetle population, especially as time progresses.


2. How much of an effect did setting up traps have on the beetle population of 150 years?  
Setting up traps that catch 5% of the second- and third-year beetles each year has a dramatic long-term effect. Over 150 years, the population with traps would either stabilize at a much lower number or even decline, while the population without traps would continue to grow. This demonstrates that even a small increase in adult mortality can prevent population outbreaks and may eventually lead to population decline or eradication.
