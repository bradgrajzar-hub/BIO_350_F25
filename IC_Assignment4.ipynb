Question 3: At what point does alpha need to be for both to coexist 

For both species to coexist under the Lotka–Volterra competition model, the competition coefficients (𝛼) must be less than 1. This means each species must limit its own growth more strongly than it limits the growth of the other species. In our case, 𝛼NM=0.25 meets this condition, but 𝛼MN=4 does not, so the threatened species cannot coexist with the colonizer in the long term.
