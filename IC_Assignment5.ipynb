```python
import matplotlib.pyplot as plt

def sir_model(beta, d, S0, I0, R0, days=30):
    S = [S0]
    I = [I0]
    R = [R0]


    for t in range(days):
        St = S[-1]
        It = I[-1]
        Rt = R[-1]

        new_S = St - beta * St * It
        new_I = It + beta * St * It - d * It
        new_R = Rt + d * It

        S.append(new_S)
        I.append(new_I)
        R.append(new_R)

    return S, I, R



beta = 0.001   # infection rate
d = 0.5        # death rate
S0 = 999
I0 = 1
R0 = 0
days = 30

S, I, R = sir_model(beta, d, S0, I0, R0, days)

print("Day | Susceptible | Infected | Removed")
for day in range(days + 1):  # include day 0
    print(f"{day:3d} | {S[day]:11.2f} | {I[day]:8.2f} | {R[day]:8.2f}")


plt.figure(figsize=(8,5))
plt.plot(range(days+1), S, label="Susceptible (S)")
plt.plot(range(days+1), I, label="Infected (I)")
plt.plot(range(days+1), R, label="Removed/Dead (R)")
plt.xlabel("Day")
plt.ylabel("Number of Foxes")
plt.title("SIR Model of Rabies in Foxes (30 days)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

Beta variable - What happens when it is changed up or down? What are reasons it may change? 
If the Beta variable is increased, the infection spreads more quickly through the population because each infected fox is more likely to transmit rabies to a susceptible one. A higher beta leads to a faster rise in infected individuals, a sharper epidemic peak, and more total deaths. If beta is lowered, the disease spreads more slowly or may not spread at all, resulting in fewer infections and deaths. In real populations, beta could change due to fox density, how often they encounter each other, their behavior, or environmental factors that affect contact rates.

d variable - What happens when the death rate is changed? Why might the rate change, or why might it be high/low?
 If the death rate (d) is higher, infected foxes die faster, which shortens the time they can spread the disease. This reduces the total number of secondary infections, sometimes preventing the epidemic from spreading widely. If d is lower, infected foxes live longer while contagious, giving them more time to infect others, which increases the spread and can wipe out much of the population. In nature, the death rate may vary depending on how rabies affects individuals physiologically, differences in immune response, or other health factors. A higher d means rabies kills very quickly, while a lower d means foxes survive longer but remain infectious.
