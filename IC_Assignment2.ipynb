- Calculating how quickly duckweed will cover a pond


```python
import matplotlib.pyplot as plt
```

```python
def N_exp(N_0, p, time): 
    # Exponential growth model
    # Save N_values into a list
    N_values = [N_0] 
    # Save initial population
    current_N = N_0 
    for _ in range(time):
        current_N = current_N + p * current_N 
        # rund to 3 decimal places 
        current_N = round(current_N, 3)
        N_values.append(current_N) 
    return N_values
```

```python
#Run the function with 35days and p = 0.4 with an initial population of 0.01
result = N_exp(0.01, 0.4, 35)
print(result)
```

```python
# Plot the results with days on the x-axis and population size on the y-axis
import matplotlib.pyplot as plt
days = list(range(36))  # 0 to 35 days
plt.plot(days, result, marker='o')
plt.xlabel('Days')
plt.ylabel('Duckweed Population Size')
plt.title('Duckweed Population Growth Over Time')
plt.grid()
plt.show()
```
