```python
initial_plants = 12
max_plants = 400
mortality = 0.4
years = 0
plants = initial_plants

plant_counts = [plants]
year_list = [years]

# Copilot helped with this loop
while plants < max_plants:
    survivors = int(plants * (1 - mortality))
    new_plants = min(survivors, max_plants - survivors)
    plants = survivors + new_plants
    years += 1
    plant_counts.append(plants)
    year_list.append(years)

import matplotlib.pyplot as plt

plt.plot(year_list, plant_counts, marker='o')
plt.xlabel('Years')
plt.ylabel('Number of Plants')
plt.title('Plant Recolonization Over Time')
plt.grid()
plt.show()

print(f"{years} years")
```
